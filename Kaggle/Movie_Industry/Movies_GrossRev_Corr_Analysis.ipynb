```python
# Import libraries

import pandas as pd # data manipulation
import seaborn as sns # data visualization
import numpy as np # numerical operations

import matplotlib 
import matplotlib.pyplot as plt # data visualization

#Matplotlib setup
plt.style.use('ggplot') #applies a clean, high-contrast visual theme.
from matplotlib.pyplot import figure

#ensures that plots are displayed in Jupyter notebooks
%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data
df = pd.read_csv(r'/Users/jincheng/Desktop/Data_Science/Kaggle/Movie_Industry/movies.csv')
```

```python
# Look at the first few rows of data
df.head(10)
```

```python
# Check if there is any missing data

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))

#df.info()          # Check data types and null values
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change the data type of columns

#df['budget'] = df['budget'].astype('int64')

#df['gross'] = df['gross'].astype('int64')

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
# Create correct Year column - since there are two columns about year (released & year)

#df['yearcorrect'] = df['released'].astype(str).str[:4]
#df['yearcorrect'] = df['released'].astype(str).str.slice()
df['yearcorrect'] = df['released'].astype(str).str.extract(r'(\d{4})')
df.head(10)
```

```python
# Sort the movie list by 'gross'
df = df.sort_values(by=['gross'], inplace=False,ascending=False)
df.head(10)
```

```python
# Set all of the displays to show max num of rows of data

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplcates

df = df.drop_duplicates()
```

```python
# Budget high correlation
# company high correlation
```

```python
# scatter plot with budget vs. gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Let's start looking at correlation
```

```python
# Conclusion: High correlation b/t budget and gross
# I was right.
```

```python
correlation_matrix = df.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company
df.head()
```

```python
# Numerize columns of features with non-numeric types

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head(10)
```

```python
df_numerized.corr()
```

```python
corr_matrix = df_numerized.corr() #method='pearson'

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Unstack the matrix

corr_pairs = corr_matrix.unstack() # each feature vs. all of the features, more clear
corr_pairs
```

```python
# Sort the 1-1 pairs corr_pairs

sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs.head(10)
```

```python
# Feature selection with corr > 0.5

high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

```python
# Feature selection: Filter out self-correlations

high_corr = high_corr[high_corr<1.0]

high_corr.head(10)

#Question: how to remove symmetric duplicate pairs - by masking the upper triangle of the correlation matrix
```
